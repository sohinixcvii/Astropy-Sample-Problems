1) Open the fits image of 3C273   
2) Show the image in a color plot and overlay it with contours   
3) Find out from the image header, the size of the image and the coordinate of the center pixel

**Downloaded 3C273 fits file from:** http://dc.zah.uni-heidelberg.de/sasmirala/q/prod/qp/3C%20273

```python
#Importing necessary packages
import numpy as np
from astropy.io import fits as f
import matplotlib.pyplot as plt
```

```python
#Storing data into an array to display image
imd=f.getdata("3C273.fits")
#Showing the shape of the array indicating the dimension of the image
print("Shape of the fits image array is: ",imd.shape)

#Opening data file to check the header information and displaying it
fh=f.open("3C273.fits",memap=True)
hd=f.PrimaryHDU()
print("Header information: \n")
hd.header
```

```python
#Plotting the image with a colorbar
plt.imshow(imd,cmap='gray')
plt.colorbar()

#Overlaying a contour map
xlist=np.linspace(0.0,50.0,50)
ylist=np.linspace(0.0,50.0,50)
X,Y=np.meshgrid(xlist,ylist)
plt.contour(X,Y,imd,colors="gray")
```

```python
#Printing the total number of pixels in the image
print("Size of the image is: ",np.size(imd)," pixels")
```

```python
#Printing the minimum, maximum and mean values of the pixels
print("Min: "+str(np.min(imd)))
print("Max: "+str(np.max(imd)))
print("Mean: "+str(np.mean(imd)))
```

```python
#Displaying the location of the central (also the brightest) pixel
m=np.max(imd)
print("Position of the brightest central pixel is: ",np.where(imd==m))
```
